# 
Outlier Removal (Quartile Method)

Interquartile range measures the spread of the middle half of our data.

Formula: Q3 - Q1

where Q1- quartile 1 and Q3- quartile 3

lower limit of the data is given by Q1-1.5*IQR

upper limit of the data is given by Q3+1.5*IQR


```python
#boxplot for visualizing for outliers
fig,ax=plt.subplots(nrows=1,ncols=3,figsize=(25,5))
sns.boxplot(df["trip_duration"],ax=ax[0])
sns.boxplot(df['passenger_count'],ax=ax[1])
sns.boxplot(df['distance'],ax=ax[2])
```

```python


#finding differnt quarters of trip_duration column
trip_duration_Q1=df['trip_duration'].quantile(0.25)
print('first quartile value ie 25th percentile of trip duration:',trip_duration_Q1)
trip_duration_Q2=df['trip_duration'].quantile(0.50)
print('second quartile value ie 50th percentile of trip duration:',trip_duration_Q2)
trip_duration_Q3=df['trip_duration'].quantile(0.75)
print('third quartile value ie 75th percentile of trip duration:',trip_duration_Q3)
     
```

```python


# calculate interquartile range
IQR=trip_duration_Q3-trip_duration_Q1
print('IQR:',IQR)
trip_duration_lower_limit=trip_duration_Q1-1.5*IQR
trip_duration_upper_limit=trip_duration_Q3+1.5*IQR
print('The lower limit of trip duration:',trip_duration_lower_limit)
print('The upper limit of trip duration:',trip_duration_upper_limit)
```

```python
# Calculate the IQR
Q1 = df['trip_duration'].quantile(0.25)
Q3 = df['trip_duration'].quantile(0.75)
IQR = Q3 - Q1

# Define the upper and lower bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove outliers
df = df[(df['trip_duration'] >= lower_bound) & (df['trip_duration'] <= upper_bound)]
```

# MACHINE LEARNING MODELS


```python
import math


# define a  function to calculate evaluation metrics
def evaluation_metrics (x_train,y_train,y_predicted):
  MSE=round(mean_squared_error(y_true=y_train, y_pred=y_predicted),4)
  RMSE=math.sqrt(MSE)
  R2_score=r2_score(y_true=y_train, y_pred=y_predicted)
  Adjusted_R2_score=1-((1-( R2_score))*(x_train.shape[0]-1)/(x_train.shape[0]-x_train.shape[1]-1))

  print("Mean Squared Error:",MSE,"Root Mean Squared Error:", RMSE)
  print("R2 Score :",R2_score,"Adjusted R2 Score :",Adjusted_R2_score)
  # plotting actual and predicted values
  #Plotting Actual and Predicted Values
  plt.figure(figsize=(18,6))
  plt.plot((y_predicted)[:100], color='red') 
  plt.plot(np.array(y_train)[:100], color='green')
  plt.legend(["Predicted","Actual"])
  plt.title('Actual and Predicted Time Duration')

  #return(MSE,RMSE,R2_score,Adjusted_R2_score)
     
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
import xgboost as xgb
from sklearn.metrics import mean_squared_error, r2_score

# Load the dataset
df = pd.read_csv('train.csv')

# Define the features and target variable
features = ['vendor_id', 'passenger_count', 'pickup_longitude', 'pickup_latitude',
            'dropoff_longitude', 'dropoff_latitude']
categorical_features = ['store_and_fwd_flag']

target = df['trip_duration']

# Perform one-hot encoding on the categorical features
df_encoded = pd.get_dummies(df[features + categorical_features])

# Create the Linear Regression model
linear_regression_model = LinearRegression()
# Train the Linear Regression model on the entire dataset
linear_regression_model.fit(df_encoded, target)
# Generate predictions on the entire dataset
linear_regression_predictions = linear_regression_model.predict(df_encoded)
# Evaluate the performance of the Linear Regression model
linear_regression_mse = mean_squared_error(target, linear_regression_predictions)
linear_regression_rmse = mean_squared_error(target, linear_regression_predictions, squared=False)
linear_regression_r2 = r2_score(target, linear_regression_predictions)

decision_tree_r2 = r2_score(target, decision_tree_predictions)



# Print the evaluation metrics
print('Linear Regression')
print('Mean Squared Error:', linear_regression_mse)
print('Root Mean Squared Error:', linear_regression_rmse)
print('R2 Score:', linear_regression_r2)
print()


```

```python
# Create the Decision Tree Regression model
decision_tree_model = DecisionTreeRegressor()
# Train the Decision Tree Regression model on the entire dataset
decision_tree_model.fit(df_encoded, target)
# Generate predictions on the entire dataset
decision_tree_predictions = decision_tree_model.predict(df_encoded)
# Evaluate the performance of the Decision Tree Regression model
decision_tree_mse = mean_squared_error(target, decision_tree_predictions)
decision_tree_rmse = mean_squared_error(target, decision_tree_predictions, squared=False)
decision_tree_r2 = r2_score(target, decision_tree_predictions)

print('Decision Tree Regression')
print('Mean Squared Error:', decision_tree_mse)
print('Root Mean Squared Error:', decision_tree_rmse)
print('R2 Score:', decision_tree_r2)
print()
```

```python

# Create the XGBoost Regression model
xgb_model = xgb.XGBRegressor()
# Train the XGBoost Regression model on the entire dataset
xgb_model.fit(df_encoded, target)
# Generate predictions on the entire dataset
xgb_predictions = xgb_model.predict(df_encoded)
# Evaluate the performance of the XGBoost Regression model
xgb_mse = mean_squared_error(target, xgb_predictions)
xgb_rmse = mean_squared_error(target, xgb_predictions, squared=False)
xgb_r2 = r2_score(target, xgb_predictions)

print('XGBoost Regression')
print('Mean Squared Error:', xgb_mse)
print('Root Mean Squared Error:', xgb_rmse)
print('R2 Score:', xgb_r2)
```
